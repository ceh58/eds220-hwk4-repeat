### About:
- Purpose: 

The purpose of this assignment is to use false color imagery to visualize the impact (the burn scar) of the Thomas Fire in 2017. 

- Highlights:

This assignment explores assigning infrared bands to visible colors and mapping shapefiles over the resulting images. Necessary steps include cleaning rasters, filtering geo-dataframes, and matching Coordinate Reference Systems (CRSs).   

- Dataset description:

Landsat Data:
Pre-processed, simplified collection of bands (red, green, blue, nir, swir) from Landsat Collection 2 Level-2 (collected by Landsat 8 satellite). 

Fire Perimeters Data:
Open-source data containing information from the spatial distrubtion of past fires in California published by the State of California and downloaded as a shapefile. 

- References:

Landsat Data:
Microsoft Open Source, Matt McFarland, Rob Emanuele, Dan Morris, & Tom Augspurger. (2022). microsoft/PlanetaryComputer: October 2022 (2022.10.28). Zenodo. https://doi.org/10.5281/zenodo.7261897

Fire Perimeters Data:
State of California, Kimberly Wallin. (2024). CAL FIRE: May 2024 (2024.05.14). https://catalog.data.gov/dataset/california-fire-perimeters-all-b3436

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import xarray as xr
import rioxarray as rioxr
import numpy as np
```

## 3. True Color Image

```python
# a. Import data
fp = os.path.join("/courses/EDS220/data/hwk4_landsat_data", "landsat8-2018-01-26-sb-simplified.nc")
landsat = rioxr.open_rasterio(fp)
landsat.head(3)
```

```python
# b. Preliminary exploration
print('dims:', landsat.dims)
print('attrs:', landsat.attrs)
landsat
```

```python
# c. Drop the 'band' dimension
landsat = landsat.squeeze().drop_vars('band')
```

```python
# d. Select 'red', 'green', and 'blue' variables and plot
xr.Dataset(landsat[['red', 'green', 'blue']]).to_array().plot.imshow()
```

```python
# e. Adjust the scale for a true color plot
xr.Dataset(landsat[['red', 'green', 'blue']]).to_array().plot.imshow(robust = True)
```

## 4. False color image

```python
# Create a false color image
xr.Dataset(landsat[['swir22', 'nir08', 'red']]).to_array().plot.imshow(robust = True)
```

## 5. Map

```python
# Import Thomas Fire shapefile
thomas_fire = gpd.read_file(os.path.join("data", "thomas_fire.shp"))
```

```python
# Make sure CRSs match
if thomas_fire.crs == landsat.rio.crs:
    print("CRSs match!")
else:
    landsat = landsat.rio.reproject(thomas_fire.crs)
    assert landsat.rio.crs == thomas_fire.crs
    print("We matched the CRSs!")
```

```python
# a. Map the false color image with the fire perimeter
landsat_aspect_ratio = landsat.rio.width/landsat.rio.height

fig, ax = plt.subplots(figsize = (6, 6*landsat_aspect_ratio))
ax.axis("off")

xr.Dataset(landsat[['swir22', 'nir08', 'red']]).to_array().plot.imshow(ax = ax,
                                                                       robust = True)

thomas_fire.plot(ax = ax,
                 color = "none",
                edgecolor = "black",
                legend = True)

ax.set_title("Thomas Fire Scar (2017)", fontsize=14, fontweight='bold')

plt.show()
```
